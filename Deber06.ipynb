### Polinomio de Taylor para la primera y segunda funcion 

```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as sp

def f_original(x): 
    return  1 / (25 * x**2 + 1)

def f_taylor(x):
    return 625*x**4 - 25*x**2 + 1

x = np.linspace(-0.2,0.2,100) 
y_original = f_original(x)
y_taylor = f_taylor(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_taylor, label='Polinomio de Taylor de orden n=5')
plt.plot(x, y_original, label='Funci√≥n Original 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title( r'Aproximaciones de Taylor $\frac{1}{25x^2 + 1}$')
plt.legend()
plt.grid()
plt.show()

```

```python

x = sp.Symbol('x')
f = sp.atan(x)
x0 = 1

n = 5
taylor_series = sp.series(f, x, x0, n+1).removeO()


taylor_func = sp.lambdify(x, taylor_series, 'numpy')
original_func = sp.lambdify(x, f, 'numpy')

# Crear valores de x para graficar
x_vals = np.linspace(-2, 4, 500)
y_original = original_func(x_vals)
y_taylor = taylor_func(x_vals)

# Graficar la funci√≥n original y la aproximaci√≥n de Taylor
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_original, label="Funci√≥n original", color="blue")
plt.plot(x_vals, y_taylor, label=f"Polinomio de Taylor de orden {n}", color="black", linestyle="--")
plt.axvline(x=x0, color="gray", linestyle=":", label=f"Punto de expansi√≥n ($x_0={x0}$)")
plt.title( r'Aproximaciones de Taylor: $\arctan(x)$')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

### Polinomio de Lagrange para la primera y segunda funcion 

```python
def f_lagrange(x):
    return -5*x**2 + 1

def f_original(x): 
    return  1 / (25 * x**2 + 1) 


original_points_x = [0, -0.4, 0.4]
original_points_y = [1, 0.2, 0.2] 

x_vals = np.linspace(-0.4, 0.4, 500)
y_original = f_original(x_vals)
y_lagrange = f_lagrange(x_vals) 

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_original, label="Funci√≥n original", color="black")
plt.plot(x_vals, y_lagrange, label="Polinomio de Lagrange manual (-5x^2 + 1)", color="blue", linestyle="--")
plt.scatter(original_points_x, original_points_y, color="red", label="Puntos originales", zorder=5)
plt.title("Interpolaci√≥n de Lagrange ")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()

```

```python
# Funci√≥n original
def f(x):
    return np.arctan(x)

# Aproximaci√≥n de Taylor de segundo orden
def lagrange_approx(x):
    pi_4 = np.pi / 4
    return pi_4 + 0.5 * (x - 1) - 0.25 * (x - 1)**2

# Valores de x
x_values = np.linspace(0, 2, 400)
y_values_original = f(x_values)
y_values_approx = lagrange_approx(x_values)

# Gr√°fica
plt.plot(x_values, y_values_original, label="arctan(x)", color="blue")
plt.plot(x_values, y_values_approx, label="Aproximaci√≥n de Lagrange", color="red", linestyle='--')
plt.scatter(1, f(1), color="green", zorder=5)  
plt.title("Aproximaci√≥n de Lagrange de segundo orden de arctan(x) ")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
